```
import math
import matplotlib.pyplot as plt
import numpy as np
```

```
max_val = 10000
data_size = 100
n = 5

X = np.random.uniform(0, max_val, (data_size, 2))

_ = plt.plot(X[:, 0], X[:, 1], 'bo')
```

```
# Training

steps = 100
lr = 0.1
W = np.random.uniform(0, max_val, (n, 2))

plt.subplot(2, 1, 1)
plt.plot(W[:, 0], W[:, 1], 'bo') # before training

for _ in range(steps):  
  
  for x in X:
    d_min = math.inf
    i_min = -1

    for i in range(len(W)):
      d = np.linalg.norm(x - W[i]) # euclidean distance

      if(d <= d_min):
        d_min = d
        i_min = i

    W[i_min] += lr*(x - W[i_min])

plt.subplot(2, 1, 2)
_ = plt.plot(W[:, 0], W[:, 1], 'bo') # after training
```

```
# Prediction

C = ['b', 'g', 'r', 'c', 'm', 'y', 'k'] # colors
C = C[:n]

for x in X:
  d_min = math.inf
  i_min = -1

  for i in range(len(W)):
    d = np.linalg.norm(x - W[i]) # euclidean distance

    if(d < d_min):
      d_min = d
      i_min = i
      
  plt.plot(x[0], x[1], 'bo', c=C[i_min])
```
